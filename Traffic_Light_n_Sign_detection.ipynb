```python
from IPython.display import HTML
import base64
import os

# Path to the original video file
save_path = 'path/to/original/video.mp4'

# Compress the video using ffmpeg
compressed_path = os.path.splitext(save_path)[0] + '_compressed.mp4'
!ffmpeg -i $save_path -vcodec libx264 $compressed_path

# Read the compressed video file as binary
with open(compressed_path, 'rb') as file:
    video_data = file.read()

# Encode the video data in Base64
encoded_data = base64.b64encode(video_data).decode()

# Create the data URL for the video
data_url = "data:video/mp4;base64," + encoded_data

# Display the video in the notebook
HTML(f'<video width=400 controls><source src="{data_url}" type="video/mp4"></video>')
```
